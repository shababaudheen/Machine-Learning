## Classification

```python
from sklearn import datasets

iris = datasets.load_iris()
iris
```

```python
x = iris.data
y = iris.target
```

Since all features have the same unit (i.e., cm), scaling is unnecessary.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
lm
```

```python
lm.fit(x_train, y_train)
```

```python
y_pred = lm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

Accuracy = accuracy_score(y_test, y_pred)
Accuracy
```

```python
sample = [[6, 3.5, 5.6, 2.], [5, 3.7, 1, 3], [4.8, 3. , 1.4, 0.1]]
predict_answer = lm.predict(sample) 
predict_answer
```

## Clustering

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = ["Sepal_length", "Sepal_width", "Petal_length", "Petal_width"])
df
```

```python
features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
```

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters = 3, random_state = 42)
kmeans
```

```python
kmeans.fit(df[features])
```

```python
y_pred = kmeans.predict(df[features])
y_pred
```

```python
df["Species"] = iris.target_names[y_pred]
df["Cluster"] = y_pred
```

```python
sns.pairplot(df, hue = "Species")
plt.show()
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
df1 ,df2, df3 
```

```python
plt.scatter(df1.Petal_length, df1.Sepal_length, color='green')
plt.scatter(df2.Petal_length, df2.Sepal_length, color='red')
plt.scatter(df3.Petal_length, df3.Sepal_length, color='black')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='blue', label='centroid') 

plt.xlabel('Petal length')
plt.ylabel('Sepal_length')
plt.legend()

plt.show()
```

### Elbow method

```python
WCSS = []
k_range = range(1, 10)

features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
for k in k_range: 
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(df[features])
    WCSS.append(kmeans.inertia_) 
WCSS
```

```python
plt.plot(k_range, WCSS)
plt.xlabel("k")
plt.ylabel("WCSS") 
plt.title("Elbow Plot")
plt.show()
```

From this plot, it is clear that k = 2, as the elbow shape forms at k = 2.

### Silhouette Coefficient

```python
from sklearn.metrics import silhouette_score

S = []
k_range = range(2, 11)

features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
for k in k_range: 
    kmeans = KMeans(n_clusters = k)
    km_pred = kmeans.fit_predict(df[features])
    S.append(silhouette_score(df[features], km_pred)) 
S
```

```python
plt.plot(k_range, S)
plt.xlabel("Clusters")
plt.ylabel("Silhouette score") 
plt.title("Silhouette Score v/s Cluster")
plt.show()
```

Here, when cluster k = 2, the Silhouette score appears to be the highest. So the number cluster need is 2.
